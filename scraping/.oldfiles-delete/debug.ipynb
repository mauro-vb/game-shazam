```python
from googleapiclient.discovery import build
import os

api_key = os.getenv('YOUTUBE_API_KEY')


def get_ids(game: str):
    '''returns a list of video ids to call later
    in the main function'''

    # empty list for storing ids
    vid_ids = []

    # creating youtube resource object
    youtube = build('youtube', 'v3',developerKey=api_key)

    # retrieve youtube video results
    response=youtube.search().list(
    part='snippet',
    q=f'{game} gameplay'
    ).execute()
    count = 0

    # iterate video response
    while response:

		# extracting required info
		# from each result object
        count+=1
        for item in response['items']:

			# extracting ids
            id = item['id']['videoId']
            vid_ids.append(id)
            print("Got video ID successfully")

		# repeat again
        if 'nextPageToken' in response and count < 2:
            response = youtube.search().list(
                    part='snippet',
                    q=f'{game} gameplay'
                    ).execute()
        else:
            break
    return vid_ids
```

```python
import cv2
print(cv2.__version__) # WE CAN HAVE DIFFERENT VERSIONS!

def extract_frames(pathIn:str, pathOut:str, ident:str):
    '''receives
    pathIn video file path
    pathOut to screenshots
    id from the video id list
    creates image files in pathOut folder'''
    
    print("got into extract frames succesfully")
    count = 0
    vidcap = cv2.VideoCapture(pathIn)
    success,image = vidcap.read()
    success = True
    while success:
        try:
            print("while ok")
            vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*5000)) # each 5 seconds
            success,image = vidcap.read()
            print ('Read a new frame: ', success)
            cv2.imwrite(pathOut + f"{ident}_frame%d.jpg" % count, image) # save frame as JPEG file
            count = count + 1
        except:
            continue
```

```python
from pytube import YouTube

game = "fifa"

id_list = get_ids(game)
id_list = id_list[:2]

print("--------------")
print("FINISHED VIDEO ID LIST")

for each in id_list:

    try:
        print(f"Trying video {each}\n")
        yt = YouTube(f'https://www.youtube.com/watch?v={each}')
        yt.streams.filter(
            progressive=True,
            file_extension='mp4').order_by('resolution').asc().first().download(
                output_path = 'temp_videos',filename=f'{game}_{each}.mp4')


        print("Extracting frames...\n")
        extract_frames(f'temp_videos/{game}_{each}.mp4', f'screenshots/{game}', ident=each)

    except:
        continue
```
